```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
cities=pd.read_csv('california_cities.csv')
cities.head(20)
```

```python
latitude, longitude=cities["latd"],cities["longd"]
population,area=cities["population_total"],cities["area_total_km2"]

# Assuming you have defined the variables longitude, latitude, population, and area before this code snippet

sns.set()
plt.scatter(longitude, latitude, label=None, c=np.log10(population), cmap='viridis', s=area, linewidths=0, alpha=0.5)
plt.gca().set_aspect('equal', adjustable='box')  # Correct way to set aspect ratio
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3, 7)

for area_value in [100, 300, 500]:
    plt.scatter([], [], c='k', alpha=0.3, s=area_value, label=str(area_value) + ' km$^2$')

plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='City Areas')
plt.title("Area and Population of California Cities")
plt.show()
```
